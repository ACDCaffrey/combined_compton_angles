Load the two slices, Cs-137 and Co-60

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.fftpack as fp
```

```python
co_slice = pd.read_csv("/Users/adamcaffrey/Desktop/co.txt", header = None).values
cs_slice = pd.read_csv("/Users/adamcaffrey/Desktop/cs.txt", header = None).values
both = pd.read_csv("/Users/adamcaffrey/Desktop/both.txt", header = None).values
```

Plot the individual reconstructions, combined when ran separately and combined when ran together

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(figsize = (15,5), ncols = 4)
p1 = ax1.imshow(co_slice)
p2 = ax2.imshow(cs_slice)
p3 = ax3.imshow(cs_slice+co_slice)
p4 = ax4.imshow(both)
ax1.set_title("Co projections")
ax2.set_title("Cs projections")
ax3.set_title("Reconstructed individually")
ax4.set_title("Reconstructed together")
cbar1 = fig.colorbar(p1, ax = ax1).minorticks_on()
cbar2 = fig.colorbar(p2, ax = ax2).minorticks_on()
cbar3 = fig.colorbar(p3, ax = ax3).minorticks_on()
cbar3 = fig.colorbar(p4, ax = ax4).minorticks_on()
plt.show()
```

Need to use a high-pass filter

```python
together_slice = (cs_slice+co_slice)[150:400,0:250]
plt.imshow(together_slice)
```

```python
F1 = fp.fft2((together_slice).astype(float))
F2 = fp.fftshift(F1)
(w,h) = together_slice.shape
half_w, half_h = int(w/2),int(h+2)
n = 300
F2[half_w-n:half_w+n+1,half_h-n:half_h+n+1] = 0
im1 = fp.ifft2(fp.ifftshift(F2)).real

fig, (ax1, ax2) = plt.subplots(figsize = (8,5), ncols = 2)
p1 = ax1.imshow(im1)
p2 = ax2.imshow(cs_slice+co_slice)
cbar1 = fig.colorbar(p1, ax = ax1).minorticks_on()
cbar2 = fig.colorbar(p2, ax = ax2).minorticks_on()
ax1.set_title("after filter")
ax2.set_title("before filter")
plt.show()

```
